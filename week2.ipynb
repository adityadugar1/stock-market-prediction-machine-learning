## 1.5 Function to calculate the average price over the last N days of stock price data where
# N-user-defined parameter

```python
df=df.set_index('Date')
df
```

```python
def avg_price(N):
    new = str(N)+'D'
    new_df=df.last('{}'.format(new))
    print(pd.Series.mean(new_df['Close Price']))
```

# Creating another function for profit loss calculations over N days.

```python
def prof_loss_per(N):
    string = str(N)+'D'
    new_df=df.last('{}'.format(string))
    c = (new_df['Close Price'].iloc[0]-new_df['Close Price'].iloc[-1])*100/(new_df['Close Price'].iloc[0])
    if c<0:
        print(str(c*100)+'% loss')
    else:
        print(str(c*100)+'% profit')
```

# 1.6 adding a column for daily change in percentages using pandas function pct_change().

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change()
```

```python
df = df.reset_index()

df = df.drop(0,axis=0)
df
```

```python
trend = list()
```

```python
for i in range(1,494):
    if df['Day_Perc_Change'][i]>-0.5 and df['Day_Perc_Change'][i]<0.5:
        trend.append('Slight or No Change')
    if df['Day_Perc_Change'][i]>0.5 and df['Day_Perc_Change'][i]<1:
        trend.append('Slight Positive')
    if df['Day_Perc_Change'][i]>-1 and df['Day_Perc_Change'][i]<-0.5:
        trend.append('Slight Negative')
    if df['Day_Perc_Change'][i]>1 and df['Day_Perc_Change'][i]<3:
        trend.append('Positive')
    if df['Day_Perc_Change'][i]>-3 and df['Day_Perc_Change'][i]<-1:
        trend.append('Negative')
    if df['Day_Perc_Change'][i]>3 and df['Day_Perc_Change'][i]<7:
        trend.append('Among Top gainers')
    if df['Day_Perc_Change'][i]>-7 and df['Day_Perc_Change'][i]<-3:
        trend.append('Among Top Losers')
    if df['Day_Perc_Change'][i]>7:
        trend.append('Bull Run')
    if df['Day_Perc_Change'][i]<-7:
        trend.append('Bear Drop')
```

```python
df['Trends']=trend
```

```python
df
```

# 1.8 Finding average and median values of the column Total Traded Quantity

```python
df4=df.groupby('Trends')
for name,group in df4:
    print(name)
    print("Median : " + str(pd.Series.median(group['Total Traded Quantity'])))
    print("Average : " + str(pd.Series.mean(group['Total Traded Quantity'])))
```
