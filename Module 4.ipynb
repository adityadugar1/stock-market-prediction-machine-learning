#4.1 

```python
df=pd.read_csv("FORTIS.csv")
df.head()
```

```python
#Creating bollinger bands
```

```python
rolling_standard=df['Close Price'].rolling(7).std()
df['Rolling mean']=df['Close Price'].rolling(7).std()
df['Upper Band']=df['Close Price'].rolling(7).mean() + rolling_standard*2
df['Lower Band']=df['Close Price'].rolling(7).mean() - rolling_standard*2
```

```python
df
```

# Filling up the bollinger columns

```python
df["Rolling mean"].fillna(df['Rolling mean'].mean(),inplace=True)
```

```python
df
```

```python
df['Upper Band'].fillna(df['Upper Band'].mean(),inplace=True)
```

```python
df['Lower Band'].fillna(df['Lower Band'].mean(),inplace=True)
```

```python
df.head()
```

# Defining the Call column

```python
#we will define call column as a nan valued numpy array and define its data type input as string.
```

```python
df['Call']=np.nan
```

```python
df['Call'].astype('str')
```

```python
for i in range(len(df)):
    if df['Close Price'][i]<df['Lower Band'][i]:
        df['Call'][i]='Buy'
    elif df['Close Price'][i]>df['Lower Band'][i] and df['Close Price'][i]<df['Rolling mean'][i]:
        df['Call'][i]='Hold Buy/Liquidate Short'
    elif df['Close Price'][i]>df['Rolling mean'][i] and df['Close Price'][i]<df['Upper Band'][i]:
        df['Call'][i] = 'Hold Short/Liquidate Buy'
    elif df['Close Price'][i]>df['Upper Band'][i]:
        df['Call'][i]='Short'
```

```python
df
```

```python
#creating input and output columns
```

```python
X=df[['Close Price','Rolling mean','Upper Band','Lower Band']]
```

```python
X
```

# 4.2 Problem

```python
#Importing the stock data of HDFC
```

```python
df_new=pd.read_csv("HDFC.csv")
```

```python
df_new.head()
```

```python
#creating new columns
```

```python
df_new.set_index('Date')
```

```python
df_new.dtypes
```

```python
df_new['open_close_pct_change']=(df_new['Close Price']-df_new['Open Price'])/df_new['Open Price']*100
df_new['high_low_pct_change']=(df_new['High Price']-df_new['Low Price'])/df_new['Low Price']*100
```

```python
df_new['Days pct change']=df_new['Close Price'].pct_change()*100
```

```python
df_new['Days pct change'][0] = 0.00
```

```python
df_new['Rolling Mean'] = df_new['Days pct change'].rolling(5).mean()
df_new['Rolling SD'] = df_new['Days pct change'].rolling(5).std()
df_new = df_new.dropna()
df_new
```

```python
df_new['Action']=np.nan
```

```python
df_new.head()
```

```python
df_new['Close Price']
```

```python
for i in range(len(df_new)-1):
    if df_new['Close Price'].iloc[i] < df_new['Close Price'].iloc[i+1]:
        df_new['Action'].iloc[i] = 1
    elif df_new['Close Price'].iloc[i] >= df_new['Close Price'].iloc[i+1]:
        df_new['Action'].iloc[i] = -1
df_new = df_new.dropna()
df_new
```

# Using Random Forest Classifier 

```python
X=df_new[['open_close_pct_change','high_low_pct_change','Rolling Mean','Rolling SD']]
```

```python
X.head()
```

```python
y=df_new['Action']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y = lb.fit_transform(y)
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
