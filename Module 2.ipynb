# 2.9

```python
df.reset_index(inplace=True)
```

```python
df.set_index('Date',inplace=True)
```

```python
# which can provide the datapoint
def trade_call(data,Period):
    j = next(i for i, x in enumerate(data) if x is not None)
    our_range = range(len(data))[j + Period - 1:]
    empty_list = [None] * (j + Period - 1)
    sub_result = [np.mean(data[i - Period + 1: i + 1]) for i in our_range]

    return np.array(empty_list + sub_result)
```

```python
trade_call(df['Average Price'],21)
```

```python
trade_call(df['Average Price'],34)
```

```python

tra = pd.DataFrame()
```

```python
tra['Date'] = df.iloc[:,0]
tra['smallSMA'] = trade_call(df['Average Price'],21)
tra['largeSMA'] = trade_call(df['Average Price'],34)
```

```python
tra['smallSMA'].plot(color='Red')
tra['largeSMA'].plot(color='Blue')
plt.legend()
```

```python
tra['signal']=0.0
tra['signal'][21:] = np.where(tra['smallSMA'][21:] 
                                            > tra['largeSMA'][21:], 1.0, 0.0)   

# Generate trading orders
tra['positions'] = tra['signal'].diff()

# Print `signals`
print(tra)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111,  ylabel='Price in $')
df['Average Price'].plot(ax=ax1, color='r', lw=2.)
tra[['smallSMA', 'largeSMA']].plot(ax=ax1, lw=2.)
#plot the buy plot
ax1.plot(tra.loc[tra.positions == 1.0].index, tra.smallSMA[tra.positions == 1.0],'^', markersize=10, color='red')
         
#Plot the sell signal
ax1.plot(tra.loc[tra.positions == -1.0].index, tra.smallSMA[tra.positions == -1.0],'v', markersize=10, color='g')
         
# Show the plot
plt.show()
```
